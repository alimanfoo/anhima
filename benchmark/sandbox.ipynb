```python
def intersect_positions_impl1(pos1, pos2):
    cond1 = np.in1d(pos1, pos2)
    cond2 = np.in1d(pos2, pos1)
    return cond1, cond2
```

```python
%%cython

import numpy as np
cimport numpy as np

def intersect_positions_impl2(np.ndarray[np.int64_t, ndim=1] pos1, 
                              np.ndarray[np.int64_t, ndim=1] pos2):
    cdef int i = 0
    cdef int j = 0
    cdef np.ndarray[np.uint8_t, ndim=1] cond1
    cdef np.ndarray[np.uint8_t, ndim=1] cond2
    cond1 = np.zeros((pos1.size, ), dtype=np.uint8)
    cond2 = np.zeros((pos2.size, ), dtype=np.uint8)
    cdef int m = pos1.size
    cdef int n = pos2.size
    
    while i < m and j < n:
        if pos1[i] < pos2[j]:
            # advance left
            i += 1
        elif pos1[i] > pos2[j]:
            # advance right
            j += 1
        else:
            # match
            cond1[i] = 1
            cond2[j] = 1
            # advance both
            i += 1
            j += 1
            
    return cond1.astype('b1'), cond2.astype('b1')
```

```python
cond1, cond2 = intersect_positions_impl2(pos1, pos2)
```

```python
_, _, af_subpop1 = anhima.af.allele_frequency(genotypes_subpop1[:, :95])
_, _, af_subpop2 = anhima.af.allele_frequency(genotypes_subpop2[:, :95])
flt = (af_subpop1 > 0) & (af_subpop1 < 1) & (af_subpop2 > 0) & (af_subpop2 < 1) & (np.abs(af_subpop1 - af_subpop2) > .4)
genotypes_subpop1_seg = genotypes_subpop1[flt]
genotypes_subpop2_seg = genotypes_subpop2[flt]
af_subpop1 = af_subpop1[flt]
af_subpop2 = af_subpop2[flt]
plt.plot(af_subpop1, af_subpop2, 'bo');
```

```python
import scipy.ndimage.filters


def allele_inheritance(genotypes, af_alt_a, af_alt_b, filter_size=0):
    af_ref_a = 1 - af_alt_a
    af_ref_b = 1 - af_alt_b
    gn = anhima.gt.as_n_alt(genotypes)
    n_variants, n_samples = gn.shape
    likelihoods = np.zeros((n_variants, n_samples, 3), dtype='f8')
    likelihood_aa = likelihoods[:, :, 0]
    likelihood_ab = likelihoods[:, :, 1]
    likelihood_bb = likelihoods[:, :, 2]
    likelihood_aa[gn == 0] = np.repeat((af_ref_a**2)[:, np.newaxis], n_samples, axis=1)[gn == 0]
    likelihood_aa[gn == 1] = np.repeat((2 * af_ref_a * af_alt_a)[:, np.newaxis], n_samples, axis=1)[gn == 1]
    likelihood_aa[gn == 2] = np.repeat((af_alt_a**2)[:, np.newaxis], n_samples, axis=1)[gn == 2]
    likelihood_ab[gn == 0] = np.repeat((af_ref_a * af_ref_b)[:, np.newaxis], n_samples, axis=1)[gn == 0]
    likelihood_ab[gn == 1] = np.repeat(((af_ref_a * af_alt_b) + (af_alt_a * af_ref_b))[:, np.newaxis], n_samples, axis=1)[gn == 1]
    likelihood_ab[gn == 2] = np.repeat((af_alt_a * af_alt_b)[:, np.newaxis], n_samples, axis=1)[gn == 2]
    likelihood_bb[gn == 0] = np.repeat((af_ref_b**2)[:, np.newaxis], n_samples, axis=1)[gn == 0]
    likelihood_bb[gn == 1] = np.repeat((2 * af_ref_b * af_alt_b)[:, np.newaxis], n_samples, axis=1)[gn == 1]
    likelihood_bb[gn == 2] = np.repeat((af_alt_b**2)[:, np.newaxis], n_samples, axis=1)[gn == 2]
    likelihoods = np.log10(likelihoods)
    if filter_size:
        likelihoods = np.apply_along_axis(np.convolve, 0, likelihoods, np.ones((filter_size,)))
        # remove edges
        likelihoods = likelihoods[filter_size//2:-1*(filter_size//2), ...]
    prediction = np.argmax(likelihoods, axis=2)
    likelihoods.sort(axis=2)
    confidence = likelihoods[:, :, 2] - likelihoods[:, :, 1]
    return prediction, confidence
```

```python
prediction, confidence = allele_inheritance(genotypes_subpop1_seg[:, 95:], af_subpop1, af_subpop2, filter_size=5)

fig, ax = plt.subplots(figsize=(18, 3))
anhima.gt.plot_discrete_calldata(prediction, colors='byr', ax=ax);

fig, ax = plt.subplots(figsize=(18, 3))
anhima.gt.plot_continuous_calldata(confidence, ax=ax, pcolormesh_kwargs=dict(cmap='Greys', vmin=0, vmax=1));

fig, ax = plt.subplots(figsize=(18, 3))
x = anhima.gt.as_012(genotypes_subpop1_seg[:, 95:])
anhima.gt.plot_discrete_calldata(x, colors='wbyr', states=(-1, 0, 1, 2), ax=ax);

fig, ax = plt.subplots(figsize=(18, 3))
x = prediction.copy()
x[confidence < .5] = -1
anhima.gt.plot_discrete_calldata(x, colors='wbyr', states=(-1, 0, 1, 2), ax=ax);

fig, ax = plt.subplots(figsize=(18, 3))
x = np.arange(af_subpop1.size) + .5
ax.plot(x, af_subpop1, 'bo-')
ax.plot(x, af_subpop2, 'ro-')
ax.set_yticks([])
ax.set_xticks([])
ax.set_xlim(0, af_subpop1.size)

# x = anhima.gt.as_012(genotypes_subpop1_seg[:, :])
# anhima.gt.plot_discrete_calldata(x, colors='wbyr', states=(-1, 0, 1, 2));

# x = anhima.gt.as_012(genotypes_subpop2_seg[:, :])
# anhima.gt.plot_discrete_calldata(x, colors='wbyr', states=(-1, 0, 1, 2));

plt.figure()
plt.hist(confidence.flatten(), bins=30);
```
