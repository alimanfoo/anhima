```python
import numpy as np
np.random.seed(1)
import scipy.stats
import random
random.seed(1)
import matplotlib.pyplot as plt
%matplotlib inline
import sys
import itertools
sys.path.insert(0, '../src')
%reload_ext autoreload
%autoreload 1
%aimport anhima.tree
%aimport anhima.sim
%aimport anhima.gt
%aimport anhima.dist
```

```python
# simulate genotypes with no relatedness
n_variants = 1000
n_samples = 100
labels = [''.join(s) for s in itertools.product('ABCDEFGHIJ', repeat=2)]
ploidy = 2
af_dist = scipy.stats.beta(a=.4, b=.6)
p_missing = 0
genotypes = anhima.sim.simulate_biallelic_genotypes(n_variants, n_samples, af_dist, p_missing, ploidy)
gn = anhima.gt.as_n_alt(genotypes)
```

```python
# simulate two sub-populations with relatedness
genotypes_subpop1 = anhima.sim.simulate_relatedness(genotypes[:, :n_samples//2], relatedness=.5, n_iter=1000)
genotypes_subpop2 = anhima.sim.simulate_relatedness(genotypes[:, n_samples//2:], relatedness=.5, n_iter=1000)
genotypes_related = np.concatenate([genotypes_subpop1, genotypes_subpop2], axis=1)
gnr = anhima.gt.as_n_alt(genotypes_related)
```

```python
# calculate pairwise distance
d, ds = anhima.dist.pairwise_distance(gn)
dr, dsr = anhima.dist.pairwise_distance(gnr)
```

```python
# build a neighbour joining tree
t = anhima.tree.nj(ds, labels=labels)
tr = anhima.tree.nj(dsr, labels=labels)
```

```python
sample_colors = (['red']*50) + (['blue']*50)
```

```python
# plot the trees
anhima.tree.plot_phylo(t, type='unrooted', tip_color=sample_colors)
```

```python
anhima.tree.plot_phylo(tr, type='unrooted', tip_color=sample_colors)
```
