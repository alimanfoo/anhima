# [``anhima.pca``](http://anhima.readthedocs.org/en/latest/pca.html) - Principal components analysis

```python
import numpy as np
np.random.seed(1)
import scipy.stats
import random
random.seed(1)
import matplotlib.pyplot as plt
%matplotlib inline
import sys
sys.path.insert(0, '../src')
%reload_ext autoreload
%autoreload 1
%aimport anhima.sim
%aimport anhima.gt
%aimport anhima.pca
```

```python
# simulate genotypes with no relatedness
n_variants = 1000
n_samples = 100
ploidy = 2
af_dist = scipy.stats.beta(a=.4, b=.6)
p_missing = 0
genotypes = anhima.sim.simulate_biallelic_genotypes(n_variants, n_samples, af_dist, p_missing, ploidy)
gn = anhima.gt.as_n_alt(genotypes)
```

```python
# simulate two sub-populations with relatedness
genotypes_subpop1 = anhima.sim.simulate_relatedness(genotypes[:, :n_samples//2], relatedness=.5, n_iter=1000)
genotypes_subpop2 = anhima.sim.simulate_relatedness(genotypes[:, n_samples//2:], relatedness=.5, n_iter=1000)
genotypes_related = np.concatenate([genotypes_subpop1, genotypes_subpop2], axis=1)
gnr = anhima.gt.as_n_alt(genotypes_related)
```

```python
# run PCA
p, t = anhima.pca.pca(gn)
pr, tr = anhima.pca.pca(gnr)
```

```python
anhima.pca.plot_components(p, t);
```

```python
anhima.pca.plot_components(pr, tr);
```
